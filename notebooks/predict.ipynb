```python
import sys
sys.path.insert(0, '..')
from bowzer.predict import Predictor
from bowzer.utils import get_target_image_dict
```

```python
target_images = get_target_image_dict()
```

```python
model = Predictor()
```

# Prediction Results:

## Anwar

Success! Anwar is classified as a Pug more than 99% of the time! Beagle trails shortly behind which is also encouraging.

```python
target_character = 'anwar'
character_image = 'headshot'
model.target_label = target_character
model.predict(target_images[target_character][character_image])
model.show_predicted_images(3, save=True)
```

```python
model.show_breed_prob_chart(save=True)
```

### Additional Angles

The prediction can vary depending on the aspect of the target image.

Below I have the same image: 'headshot_zoom' is simply 'headshot_full' but cropped a little tighter.

As you can see, the top matching image is the same but the probability distribution is different. There could be many reasons for this. For this example, the model is aligned across the top 2 breeds (Chihuahua and American Bulldog) as well as the top match image itself (the white colored chihuahua).

My hypothesis:
- The model is recognizing the white space in the background
- The model is capturing pete's smile lines and aligning with the "wrinkly" nature of American Bulldogs... no offense Paul Rudd!

In general, I think Chihuahua's have a lot of human-like features... big eyes, distinct smiles, various coloring which can align to different human hair colors. I'm not surprised that Chihuahua shows up a lot in these results.

```python
target_character = 'pete'
character_image = 'headshot_full'
model.target_label = target_character
model.predict(target_images[target_character][character_image])
model.show_predicted_images(3)
model.show_breed_prob_chart()
```
